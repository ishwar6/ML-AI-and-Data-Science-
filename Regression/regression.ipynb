```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("../data/Advertising.csv")


df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
print(df.head())

sns.scatterplot(data=df, x='total_spend', y ='sales')
```

```python
sns.regplot(data=df, x='total_spend', y ='sales')
```

```python
X = df['total_spend'] #a feature
y = df['sales']
```

```python
help(np.polyfit)
```

```python
np.polyfit(X, y, deg=1)
```

```python
spends = np.linspace(0, 500, 100)
predicted = 0.04868*spends + 4.2430
```

```python
plt.plot(spends, predicted, color='red')
```

### we can do the same for more than one degrees (x, x**2, x**3 etc)

```python
np.polyfit(X,y,3)
```

```python
pot_spend = np.linspace(0,500,100)
```

```python
predict_spend = 3.07615033e-07*pot_spend**3 -1.89392449e-04*pot_spend**2 + 8.20886302e-02*pot_spend + 2.704
```

```python
sns.scatterplot(data=df, x='total_spend', y ='sales')
plt.plot(pot_spend, predict_spend, color='red')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16,6))

axes[0].plot(df['TV'], df['sales'], 'o')
axes[0].set_ylabel('Sales')
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'], df['sales'], 'o')
axes[1].set_ylabel('Sales')
axes[1].set_title("Radio Spend")

axes[2].plot(df['newspaper'], df['sales'], 'o')
axes[2].set_ylabel('Sales')
axes[2].set_title("Newspaper Spend")
```
